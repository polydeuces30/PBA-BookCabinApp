# Scrapping Google Play Review Book Cabin App

Pengumpulan data primer dilakukan melalui teknik web scraping pada platform Google Play Store terhadap aplikasi Book Cabin. Dengan menggunakan fungsi reviews_all dari library google-play-scraper, seluruh ulasan pengguna dapat diekstraksi tanpa batasan kuantitas. Proses ini dikonfigurasi pada ID paket com.spotify.music dengan filter bahasa Indonesia ('id') dan metode pengurutan terbaru (Sort.NEWEST). Pendekatan ini memastikan dataset yang diperoleh bersifat komprehensif, relevan, dan mencerminkan kondisi terkini aplikasi sebagai landasan analisis tahap berikutnya.

Install Library

```python
!pip install google-play-scraper
from google_play_scraper import app, reviews_all, Sort
from datetime import datetime
import pandas as pd
import numpy as np
```

Informasi Aplikasi Book Cabin

```python
# 1. Mengambil detail app  BookCabin
result = app(
    'com.kabinkitatop.bookcabin',
    lang='id',
    country='id'
)

# 2. Konversi Timestamp ke Format Tanggal Manusia
dt_object = datetime.fromtimestamp(result['updated'])
tanggal = dt_object.strftime('%d %B %Y, %H:%M WIB')

# 3. Menampilkan informasi spesifik terkait update
print(f"Nama Aplikasi     : {result['title']}")
print(f"Versi Terbaru     : {result['version']}")
print(f"Terakhir Diupdate  : {tanggal}")

# 4. Menampilkan dalam bentuk DataFrame
df_detail = pd.DataFrame([result])
df_detail['updated_formatted'] = tanggal
display(df_detail[['title', 'version', 'updated_formatted', 'installs']])
```

Scrapping App

```python
# 1. Proses scrap
bookcabin_reviews_id = reviews_all(
    'com.kabinkitatop.bookcabin',
    sleep_milliseconds=0,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
)

print(f'✅ Scraping selesai! Total ulasan terkumpul: {len(bookcabin_reviews_id):,}')

# 2. Menyimpan ke DataFrame
df_bookcabin_id = pd.DataFrame(np.array(bookcabin_reviews_id), columns=['content'])
df_bookcabin_id = df_bookcabin_id.join(pd.DataFrame(df_bookcabin_id.pop('content').tolist()))

print(f'\nJumlah baris  : {df_bookcabin_id.shape[0]:,}')
print(f'Jumlah kolom  : {df_bookcabin_id.shape[1]}')

df_bookcabin_id.head(
)
```
