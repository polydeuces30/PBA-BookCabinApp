```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 1. Dataset dari slide Stanford
corpus = [
    "Pasalnya, keberadaan sejumlah lapangan padel dikeluhkan masyarakat lantaran operasionalnya mengganggu kenyamanan dan ketenangan warga.", #SinPo.id
    "Sebelumnya, sejumlah warga di Jakarta Selatan mengeluhkan operasional lapangan padel yang berlokasi tepat di sebelah hunian mereka.", #metroTV
    "Langkah ini diambil Pemerintah Provinsi (Pemprov) DKI Jakarta merespons maraknya keluhan masyarakat terkait polusi suara atau kebisingan dari aktivitas olahraga tersebut yang sering berlangsung hingga larut malam." #metroTV
]

# 2. Representasi Vektor (Hanya fokus pada kata 'lapangan' dan 'warga')
vectorizer = CountVectorizer(vocabulary=['lapangan', ' warga'])
X = vectorizer.fit_transform(corpus)

# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0 (1 John)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['lapangan', 'warga'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

**CountVectorizer:** Mengubah teks menjadi angka (matriks frekuensi kata). Di sini kita membatasi fitur hanya pada kata "hate" dan "darkness" agar sesuai dengan grafik di slide Stanford.

**Cosine Similarity vs Distance:** Fungsi ini menghasilkan nilai 0 sampai 1. Semakin mendekati 1, semakin mirip kedua dokumen tersebut.

**Analisis Hasil:** Bahwa Dokumen 2 (MLK) memiliki nilai similarity yang tinggi terhadap Dokumen 0, membuktikan bahwa meskipun jumlah katanya jauh lebih banyak, "arah" topiknya tetap serupa.

**Euclidean Distance:** Jika kita menggunakan jarak lurus, Dokumen 2 akan dianggap sangat berbeda karena jumlah kemunculan katanya (3 dan 4) jauh dari Dokumen 0 (1 dan 2).

**Aplikasi di Dunia Nyata:** Teknik ini adalah dasar dari mesin pencari (search engines) dan sistem rekomendasi teks yang sering kita gunakan sehari-hari.

```python
import numpy as np
import pandas as pd
from collections import Counter

def length(v):
    return np.sqrt((v ** 2).sum())

def cos_dist(v, w):
    # Menambahkan penanganan jika panjang salah satu vektor adalah nol
    len_v = length(v)
    len_w = length(w)
    if len_v == 0 or len_w == 0:
        return 0.0 # Atau NaN, tergantung perilaku yang diinginkan untuk vektor nol
    return 1 - (v * w).sum() / (len_v * len_w)

# Memperbaiki definisi documents agar memiliki multiple dokumen
documents = [
    "Pasalnya, keberadaan sejumlah lapangan padel dikeluhkan masyarakat lantaran operasionalnya mengganggu kenyamanan dan ketenangan warga.",
    "Sebelumnya, sejumlah warga di Jakarta Selatan mengeluhkan operasional lapangan padel yang berlokasi tepat di sebelah hunian mereka.",
    "Langkah ini diambil Pemerintah Provinsi (Pemprov) DKI Jakarta merespons maraknya keluhan masyarakat terkait polusi suara atau kebisingan dari aktivitas olahraga tersebut yang sering berlangsung hingga larut malam."
]

# Membuat ulang tf dengan documents yang sudah benar
tf = pd.DataFrame(
    [Counter(doc.split()) for doc in documents],
).fillna(0)

# Sekarang, tf memiliki multiple baris, sehingga indexing tf.loc[1] dan tf.loc[2] akan valid
# Kita bandingkan dokumen 0 dengan dokumen 1, dan dokumen 0 dengan dokumen 2
# Catatan: Karena dokumen 0, 1, dan 2 adalah string yang sangat mirip (dari dokumen asli yang sama),
# nilai cosine distance akan sangat rendah (cosine similarity akan tinggi).
# Untuk perbandingan yang lebih bermakna, dokumen seharusnya berbeda secara signifikan.
cos_dists_0_1 = cos_dist(tf.loc[0], tf.loc[1])
cos_dists_0_2 = cos_dist(tf.loc[0], tf.loc[2])
cos_dists_1_2 = cos_dist(tf.loc[1], tf.loc[2])

print(f"Cosine Distance between Document 0 and Document 1: {cos_dists_0_1}")
print(f"Cosine Distance between Document 0 and Document 2: {cos_dists_0_2}")
print(f"Cosine Distance between Document 1 and Document 2: {cos_dists_1_2}")
```
