## Langkah 2: Named Entity Recognition (NER)

Library `transformers` dari Hugging Face untuk melakukan Named Entity Recognition (NER) pada konten artikel yang telah diambil. NER akan mengidentifikasi dan mengklasifikasikan entitas

### Mengatur Hugging Face Token


```python
# Import library untuk mengelola secrets di Colab
from google.colab import userdata
import os

# Ambil HF_TOKEN dari Colab secrets
hf_token = userdata.get('HF_TOKEN')

# Set token sebagai environment variable
# Ini akan digunakan secara otomatis oleh pustaka transformers
if hf_token:
    os.environ['HF_TOKEN'] = hf_token
    print("HF_TOKEN berhasil dimuat dari Colab secrets.")
else:
    print("HF_TOKEN tidak ditemukan di Colab secrets. Peringatan mungkin masih muncul.")
```

```python
# Instal pustaka yang diperlukan untuk NER
!pip install transformers[sentencepiece]

from transformers import pipeline

# Inisialisasi pipeline NER
# Kita akan menggunakan model default yang bagus untuk bahasa Inggris.
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")

# Fungsi untuk menjalankan NER pada teks
def extract_entities(text):
    # Batasi teks karena model NER bisa memakan banyak memori/waktu untuk teks yang sangat panjang
    # Kita ambil 500 karakter pertama sebagai contoh
    entities = ner_pipeline(text[:500]) # Removed grouped_entities=True
    return entities

# Terapkan NER ke kolom 'content' pada DataFrame articles_df
# Pastikan articles_df ada dan memiliki kolom 'content'
if 'articles_df' in locals() and not articles_df.empty:
    articles_df['entities'] = articles_df['content'].apply(extract_entities)

    print("\nNER selesai. Berikut adalah entitas yang ditemukan pada beberapa artikel pertama:")
    display(articles_df[['title', 'entities']].head())
else:
    print("DataFrame 'articles_df' tidak ditemukan atau kosong. Pastikan langkah web scraping berhasil.")
```

## Langkah 3: Analisis Sentimen dengan TextBlob

Setelah mendapatkan entitas, dilanjutkan dengan menganalisis sentimen dari setiap artikel menggunakan pustaka `TextBlob`. Sentimen akan diklasifikasikan menjadi positif, negatif, atau netral.

```python
# Instal pustaka yang diperlukan untuk Analisis Sentimen
!pip install textblob
!python -m textblob.download_corpora

from textblob import TextBlob

# Fungsi untuk mendapatkan sentimen
def get_sentiment(text):
    analysis = TextBlob(text)
    # Tentukan polaritas sentimen
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Terapkan analisis sentimen ke kolom 'content'
if 'articles_df' in locals() and not articles_df.empty:
    articles_df['sentiment'] = articles_df['content'].apply(get_sentiment)

    print("\nAnalisis Sentimen selesai. Berikut adalah sentimen yang ditemukan:")
    display(articles_df[['title', 'sentiment', 'content']].head())

    # Menampilkan distribusi sentimen
    print("\nDistribusi Sentimen:")
    display(articles_df['sentiment'].value_counts())
else:
    print("DataFrame 'articles_df' tidak ditemukan atau kosong. Pastikan langkah web scraping berhasil.")
```

Sekarang, DataFrame `articles_df` memiliki kolom baru bernama `sentiment` yang menunjukkan kategori sentimen (Positif, Negatif, atau Netral) untuk setiap artikel. Anda juga dapat melihat distribusi sentimen secara keseluruhan.

## Langkah 4: Analisis TF-IDF

Setelah melakukan NER dan analisis sentimen, langkah terakhir adalah menganalisis Term Frequency-Inverse Document Frequency (TF-IDF). TF-IDF adalah teknik yang digunakan untuk mengukur seberapa penting sebuah kata dalam dokumen dalam koleksi atau korpus. Nilai TF-IDF yang tinggi menunjukkan bahwa kata tersebut sering muncul dalam dokumen tertentu tetapi jarang muncul di seluruh korpus, sehingga menandakan relevansi tinggi kata tersebut terhadap dokumen tersebut.

```python
# Instal pustaka yang diperlukan untuk TF-IDF dan visualisasi
!pip install scikit-learn matplotlib seaborn

from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Pastikan articles_df ada dan memiliki kolom 'content'
if 'articles_df' in locals() and not articles_df.empty:
    documents = articles_df['content'].tolist()

    # Inisialisasi TfidfVectorizer
    # Hapus stop words umum bahasa Inggris dan batasi fitur
    tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)

    # Hitung TF-IDF
    tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

    # Dapatkan nama fitur (kata-kata)
    feature_names = tfidf_vectorizer.get_feature_names_out()

    # Buat DataFrame dari TF-IDF matrix untuk analisis lebih lanjut
    tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

    print("\nAnalisis TF-IDF selesai. Berikut adalah matriks TF-IDF (beberapa baris dan kolom pertama):")
    display(tfidf_df.head())

    # --- Visualisasi dan Analisis TF-IDF ---

    print("\nVisualisasi Top Kata Kunci (TF-IDF) per Artikel:")
    num_top_words = 10 # Jumlah kata teratas yang ingin ditampilkan

    plt.figure(figsize=(15, int(np.ceil(len(articles_df)/2)) * 5)) # Sesuaikan ukuran figure
    for i, row in enumerate(tfidf_df.iterrows()):
        # Ambil indeks kata-kata dengan skor TF-IDF tertinggi untuk setiap artikel
        top_words_indices = row[1].nlargest(num_top_words).index
        top_words_scores = row[1].nlargest(num_top_words).values

        # Perbaikan: Konversi np.ceil ke int karena plt.subplot memerlukan integer
        plt.subplot(int(np.ceil(len(articles_df)/2)), 2, i + 1) # Layout subplot
        sns.barplot(x=top_words_scores, y=top_words_indices, legend=False) # Menghapus palette='viridis' untuk menghindari FutureWarning
        plt.title(f"Top {num_top_words} Kata Kunci Artikel {i+1} ({articles_df['title'].iloc[i][:40]}...)", fontsize=10)
        plt.xlabel('TF-IDF Score', fontsize=8)
        plt.ylabel('Kata Kunci', fontsize=8)
        plt.xticks(fontsize=7)
        plt.yticks(fontsize=7)
        plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Menyesuaikan layout

    plt.suptitle('Distribusi Top Kata Kunci TF-IDF per Artikel', fontsize=16, y=1.02)
    plt.show()

    print("\nAnalisis TF-IDF:")
    print("Dari visualisasi di atas, kita dapat mengidentifikasi kata-kata kunci yang paling menonjol di setiap artikel. Kata-kata ini memiliki nilai TF-IDF yang tinggi karena sering muncul dalam dokumen tertentu dan relatif jarang di dokumen lain dalam korpus. Hal ini membantu kita memahami fokus utama atau topik spesifik dari setiap artikel.")
    print("Misalnya, jika kata 'efficiency' sering muncul di satu artikel tetapi tidak di yang lain, itu menunjukkan bahwa artikel tersebut sangat fokus pada efisiensi pusat data atau energi. Sebaliknya, kata-kata umum seperti 'data' atau 'center' mungkin memiliki skor TF-IDF yang lebih rendah karena muncul di hampir semua artikel.")
    print("Visualisasi ini memberikan gambaran cepat tentang perbedaan tematik antar artikel dan memungkinkan identifikasi istilah-istilah yang paling relevan untuk setiap dokumen dalam konteks keseluruhan topik 'IT Data Center dan Iklim'.")

else:
    print("DataFrame 'articles_df' tidak ditemukan atau kosong. Pastikan langkah web scraping berhasil.")
```
